```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
dataset.info()
```

```python
plt.figure(figsize=(12,7))
sns.histplot(data=dataset, x='fare',bins=30, kde=True, hue='class', palette='bright')
plt.title('Distribution of Titanic Ticket Prices by Class')
plt.xlabel('Fare')
plt.ylabel('No of Passengers')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,4))
pq=sns.histplot(x='fare', bins=10, kde=False,data=dataset,hue='survived')
for i in pq.containers:
    pq.bar_label(i)
plt.show()
```

```python
sns.boxplot(x='survived',y='age',hue='sex', data=dataset)
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.title("Age Distribution of Titanic Passengers by Survival Status and Gender")
```

```python
pq=sns.barplot(x='sex', y='age', hue='survived', palette='rainbow', data=dataset)
for i in pq.containers:
    pq.bar_label(i)
plt.title('Average Age of Passengers by Gender and Survival Status in the Titanic Dataset')
plt.show()
```

```python
sns.barplot(x='sex', y='age', hue='survived', palette='rainbow', data=dataset)
plt.title('Average Age of Passengers by Gender and Survival Status in the Titanic Dataset')
plt.show()
```
